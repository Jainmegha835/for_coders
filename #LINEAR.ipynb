```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv('Salary_Data.csv')
data.head()
```

VISUALIZE RAW DATASET

```python
plt.scatter(data['YearsExperience'],data['Salary'])

plt.show()
```

```python
data.plot(x=1,y=1,kind='scatter',title='YearsExperience vs Salary',figsize=(18,10))
plt.show()
```

```python
data.plot(y=[1,1],kind="line",title="YearsExperience vs Salary",figsize=(18, 10))

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=data['YearsExperience'].values.reshape(-1,1)
y=data['Salary'].values.reshape(-1,1)
reg=LinearRegression()
reg.fit(x,y)
print("the linear reg is:y={:.5}+{:.5}x".format(reg.intercept_[0],reg.coef_[0][0]))
new_yoe=10.7
new_yoe2=11.0
print(reg.predict([[new_yoe2]]))
```

```python
prediction=reg.predict(x)
plt.scatter(data['YearsExperience'],data['Salary'])
plt.plot(data['YearsExperience'],prediction,linewidth=8,c='red')
plt.xlabel=('YearsExperience')
plt.ylabel=('Salary')
plt.show()
```
